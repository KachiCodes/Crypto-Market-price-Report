```python
#importing packages for scrapping the web for data
import requests
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
import os
```

```python
#inputing the web url from which i will scrap the data
url_link = 'https://cryptowat.ch/'
# using requests method to get the content of the url
get_html_data = requests.get(url_link).text
# passing the url content via beautiful soup parser
soup_object = soup(get_html_data, 'html.parser')
# retrieving all the links in the website
for link in soup_object.find_all('a', href=True):
    print(link.get('href'))
```

```python
#i will be exploring the https://api.cryptowat.ch/markets from which i will scrape the market price data
url_link = 'https://api.cryptowat.ch/markets'
# getting the content of our url
get_html_data = requests.get(url_link).text
# passing the url content via beautiful soup
soup_obj = soup(get_html_data, 'html.parser')
# scrape the data by creating the get_price_data function 
def get_price_data(symbol, exchange='binance', after='2017-01-01'):
    url_link = 'https://api.cryptowat.ch/markets/{exchange}/{symbol}usd/ohlc'.format(symbol=symbol, exchange=exchange)
    response = requests.get(url_link, params={
        'periods': '21600',
        'after': str(int(pd.Timestamp(after).timestamp()))
    })
    
    get_html_data = response.json()
    df = pd.DataFrame(get_html_data['result']['21600'], columns=[
        'CloseTime', 'OpenPrice', 'HighPrice', 'LowPrice', 'ClosePrice', 'volume', 'MarketCap'
    ])
    df['CloseTime'] = pd.to_datetime(df['CloseTime'], unit='s')
    df.set_index('CloseTime', inplace=True)
    return df
```

```python
# pull data from crypto for btc and eth
btc = get_price_data('btc', 'bitstamp')
eth = get_price_data('eth', 'bitstamp') 


# pull others for comparison
ada = get_price_data('ada', 'bitstamp')
xrp = get_price_data('xrp', 'bitstamp')
usdt = get_price_data('usdt', 'bitstamp')
matic = get_price_data('matic', 'bitstamp')
usdc = get_price_data('usdc', 'bitstamp')
shib = get_price_data('shib', 'bitstamp')
ftm = get_price_data('ftm', 'bitstamp')
uma = get_price_data('uma', 'bitstamp')


# displaying Ethereum marketprice data (eth) 
eth
```
